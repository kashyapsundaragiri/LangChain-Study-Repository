##  Ollama
Ollama supports embedding models, making it possible to build retrieval augmented generation (RAG) applications that combine text prompts with existing documents or other data.
### ref to this link for different embedding models: https://ollama.com/blog/embedding-models

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings= (
    OllamaEmbeddings(model="nomic-embed-text") # by default it uses llama2
)
```

```python
embeddings
```

```python
e1= embeddings.embed_documents(
    [
    "One piece is the most favourite anime",
    "Naruto is the second most favourite anime",
]
)
```

```python
e1[0]
```

```python
e2= embeddings.embed_query("What is the most favourite anime")
```

```python
e2
```

```python
# Other Embeddings
# https://ollama.com/blog/embedding-models  refer to this website for other different models

embeddings= OllamaEmbeddings(model="mxbai-embed-large")
inpu_txt= "I am the greatest of all time"
vectors= embeddings.embed_query(inpu_txt)
```
